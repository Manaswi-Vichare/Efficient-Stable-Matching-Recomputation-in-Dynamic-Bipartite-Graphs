```python
import sys

# Open the input.txt file for reading and output.txt file for writing
sys.stdin = open('input.txt', 'r')
sys.stdout = open('output.txt', 'w')

# Function to check the preference of a node 'b'
def preferenceOrder(prefer, b, a, a1):
    for i in range(N):
        if prefer[b][i] == a1:
            return True

        if prefer[b][i] == a:
            return False

# Function to find a stable marriage using the Gale-Shapley algorithm
def stableMatching(prefer, bMatching, aFree):
    freeCount = N
    maxIter = 0 #to keep track of the maximum number of interations

    while freeCount > 0 and maxIter < N * N:

        a = 0
        while a < N:
            if not aFree[a]:
                break
            a += 1

        i = 0
        while i < len(prefer[a]) and not aFree[a]:
            maxIter += 1
            b = prefer[a][i]

            if bMatching[b - N] == -1:
                bMatching[b - N] = a
                aFree[a] = True
                freeCount -= 1

            else:
                a1 = bMatching[b - N]

                if not preferenceOrder(prefer, b, a, a1):
                    bMatching[b - N] = a
                    aFree[a] = True
                    aFree[a1] = False
            i += 1
#this part prints the matchings and unmatched nodes in A and B
    print("Matchings:")

    for i in range(N):
        if bMatching[i] != -1:
            print(f"Node {bMatching[i]} (set A) --> Node {i+N} (set B)")

    print("Free nodes in set A:")
    countA = 0
    for i in range(N):
        if not aFree[i]:
            print(f"node {i}")
            countA+=1
    if countA == 0:
        print("None")

    print("Free nodes in set B:")
    countB = 0
    for i in range(N):
        if bMatching[i] == -1:
            print(f"node {i+N}")
            countB+=1
    if countB == 0:
        print("None")

    return bMatching, aFree

# Input the number of setA and setB
setA = int(input("No. of elements in set A: "))
print(setA)
setB = int(input("No. of elements in set B: "))
print(setB,"\n")
N = setA

bMatching = [-1 for i in range(setB)]
aFree = [False for i in range(setA)]
prefer = [[] for _ in range(setA + setB)]

# Input preferences for setA and setB
for i in range(len(prefer)):
    if i < setA:
        prefer[i] = list(map(int, input(f"Ordered set for node {i} (set A): ").split()))
        print(prefer[i])
    else:
        prefer[i] = list(map(int, input(f"Ordered set for node {i} (set B): ").split()))
        print(prefer[i])

matchings_complete=[]
for i in range(len(prefer)):
    for j in range(len(prefer[i])):
        matchings_complete.append([i,prefer[i][j]])

# print(matchings_complete)
print("\nStable Matchings: ")
bMatching, aFree = stableMatching(prefer, bMatching, aFree)

matchings = []
for i in range(N):
    if bMatching[i] != -1:
        matchings.append((bMatching[i], i+N))

print("\nDelete a set of nodes from the stable matching\n")
flag = True
del_edge = []
edges_del = int(input("Enter the number of edges to be deleted: "))
print(edges_del)
if edges_del == 0:
    print("\nNo edge to be deleted")
    exit()
print("")
while edges_del > 0:
    del_edge_temp = list(map(int, input("Delete Edge [set A -> set B]: ").split()))
    print(del_edge_temp)

    if ([del_edge_temp[0], del_edge_temp[1]]) in matchings_complete:

        del_edge.append(del_edge_temp)

        if (del_edge_temp[0], del_edge_temp[1]) in matchings:
            bMatching[del_edge_temp[1] - N] = -1
            aFree[del_edge_temp[0]] = False

        prefer[del_edge_temp[0]].remove(del_edge_temp[1])
        prefer[del_edge_temp[1]].remove(del_edge_temp[0])
        edges_del -= 1

    else:
        print(f"Edge {del_edge_temp[0]} --> {del_edge_temp[1]} absent")
        edges_del -= 1
        continue

print("")
bMatching, aFree = stableMatching(prefer, bMatching, aFree)
```
